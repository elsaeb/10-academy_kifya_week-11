```python
# daily percentage change
tsla_data['Daily_Percent_Change'] = tsla_data['Close'].pct_change() * 100
bnd_data['Daily_Percent_Change'] = bnd_data['Close'].pct_change() * 100
spy_data['Daily_Percent_Change'] = spy_data['Close'].pct_change() * 100

plt.figure(figsize=(14, 12))

# Tesla Daily Percentage Change
plt.subplot(3, 1, 1)
plt.plot(tsla_data['Date'],tsla_data['Daily_Percent_Change'], color='blue', label='Tesla')
plt.title("Tesla Daily Percentage Change")
plt.xlabel("Date")
plt.ylabel("Daily % Change")
plt.grid(True)

# ETF (BND) Daily Percentage Change
plt.subplot(3, 1, 2)
plt.plot(bnd_data['Date'], bnd_data['Daily_Percent_Change'], color='green', label='Vanguard BND')
plt.title("Bond Market ETF (BND) Daily Percentage Change")
plt.xlabel("Date")
plt.ylabel("Daily % Change")
plt.grid(True)

# ETF (SPY) Daily Percentage Change
plt.subplot(3, 1, 3)
plt.plot(spy_data['Date'],spy_data['Daily_Percent_Change'], color='red', label='S&P 500 ETF')
plt.title("S&P 500 ETF (SPY) Daily Percentage Change")
plt.xlabel("Date")
plt.ylabel("Daily % Change")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
rolling_window = 30

# rolling mean and rolling standard deviation
tsla_data['Rolling_Mean'] = tsla_data['Daily_Percent_Change'].rolling(window=rolling_window).mean()
tsla_data['Rolling_Std'] = tsla_data['Daily_Percent_Change'].rolling(window=rolling_window).std()

bnd_data['Rolling_Mean'] = bnd_data['Daily_Percent_Change'].rolling(window=rolling_window).mean()
bnd_data['Rolling_Std'] = bnd_data['Daily_Percent_Change'].rolling(window=rolling_window).std()

spy_data['Rolling_Mean'] = spy_data['Daily_Percent_Change'].rolling(window=rolling_window).mean()
spy_data['Rolling_Std'] = spy_data['Daily_Percent_Change'].rolling(window=rolling_window).std()

plt.figure(figsize=(14, 12))

# Tesla 
plt.subplot(3, 1, 1)
plt.plot(tsla_data['Date'], tsla_data['Rolling_Mean'], color='blue', label='Rolling Mean')
plt.plot(tsla_data['Date'], tsla_data['Rolling_Std'], color='orange', label='Rolling Std Dev')
plt.title("Tesla Rolling Mean and Standard Deviation of Daily Percentage Change")
plt.xlabel("Date")
plt.ylabel("Percentage Change (%)")
plt.legend()
plt.grid(True)

#BND
plt.subplot(3, 1, 2)
plt.plot(bnd_data['Date'], bnd_data['Rolling_Mean'], color='green', label='Rolling Mean')
plt.plot(bnd_data['Date'], bnd_data['Rolling_Std'], color='purple', label='Rolling Std Dev')
plt.title("Vanguard Total Bond Market ETF (BND) Rolling Mean and Standard Deviation of Daily Percentage Change")
plt.xlabel("Date")
plt.ylabel("Percentage Change (%)")
plt.legend()
plt.grid(True)

# S&P 500 ETF (SPY)
plt.subplot(3, 1, 3)
plt.plot(spy_data['Date'], spy_data['Rolling_Mean'], color='red', label='Rolling Mean')
plt.plot(spy_data['Date'], spy_data['Rolling_Std'], color='cyan', label='Rolling Std Dev')
plt.title("ETF SPY Rolling Mean and Standard Deviation of Daily Percentage Change")
plt.xlabel("Date")
plt.ylabel("Percentage Change (%)")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

def decompose_time_series(df, column, model='multiplicative', freq=252):
    decomposition = seasonal_decompose(df[column].dropna(), model=model, period=freq)
    return decomposition

# Tesla decomposition
decomposition_tsla = decompose_time_series(tsla_data, 'Close')
decomposition_bnd = decompose_time_series(bnd_data, 'Close')
decomposition_spy = decompose_time_series(spy_data, 'Close')

def plot_decomposition(decomposition, title):
    plt.figure(figsize=(10, 8))
    plt.subplot(4, 1, 1)
    plt.plot(decomposition.observed)
    plt.title(f"{title} - Observed")

    plt.subplot(4, 1, 2)
    plt.plot(decomposition.trend)
    plt.title(f"{title} - Trend")

    plt.subplot(4, 1, 3)
    plt.plot(decomposition.seasonal)
    plt.title(f"{title} - Seasonal")

    plt.subplot(4, 1, 4)
    plt.plot(decomposition.resid)
    plt.title(f"{title} - Residuals")
    plt.tight_layout()
    plt.show()

plot_decomposition(decomposition_tsla, "Tesla")
plot_decomposition(decomposition_bnd, "Vanguard Total Bond Market ETF (BND)")
plot_decomposition(decomposition_spy, "ETF (SPY)")
```
